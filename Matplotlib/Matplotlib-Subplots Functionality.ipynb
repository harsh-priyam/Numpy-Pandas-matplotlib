```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
 x = np.arange(0,10)
y = 2*x   
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
#axes[0].plot(x,y) 1-Dimension
axes[0][0].plot(x,y) #2-Dimension
#axes[1].plot(a,b)
axes[1][1].plot(a,b) #2-Dimension

axes[0][1].plot(x,y)
axes[1][0].plot(a,b)

axes[1][0].set_ylabel('Y LABEL 1,0')
axes[1][1].set_title('TITLE 1,1')
axes[1][1].set_xlim(2,6)


fig.suptitle("Figure Level",fontsize=16)


 #for ax in axes:
     #ax.plot(x,y)
        
 #border spacing size adjust parameter
    
plt.tight_layout() 

#fig.subplots_adjust()

#fig.subplots_adjust(wspace=0.9,hspace=0.5)

fig.set_figwidth(10)
#or
#fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(4,10),dpi=200)

fig.savefig('new_subplots.png',bbox_inches='tight')
```
